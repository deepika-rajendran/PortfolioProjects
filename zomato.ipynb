```python
#merging two tables using country code
final_df= pd.merge(df,df_country,on= 'Country Code',how= 'left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df['Country'].value_counts()
```

```python
country_names= final_df['Country'].value_counts().index
```

```python
country_val=final_df['Country'].value_counts().values
```

```python
#Top 3 country which is using zomato
plt.pie(country_val[:3],labels= country_names[:3],autopct='%1.2f%%')
```

```python
OBSERVATION: India has the highest Zomato usage, then the USA and the United Kinkdom
```

```python
#numerical variables
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

Observation: When the rating is between 4.5 and 4.9, it is exceptional; when it is between 4.4 and 4.0, it is very good.
rating is good if it is between 3.9 and 3.5.
If the rating is between 3.4 and 2.5, it is deemed average; if it is less than 2.5, it is regarded poor.

```python
ratings.head()
```

```python
plt.figure(figsize = (12,8))
sns.barplot(x= 'Aggregate rating',y='Rating count',hue='Rating color',data= ratings, palette=['Blue','red','orange','yellow','green','green'])
```

OBSERVATION:
-Number of not rated count is too high
-maximum number of rating is from 2.5-3.9



```python
#count plot
sns.countplot (x='Rating color', data= ratings, palette=['Blue','red','orange','yellow','green','green'])
```
